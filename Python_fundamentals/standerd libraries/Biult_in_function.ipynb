### Object & Attribute Handling

```python
# type(), isinstance(), id()

# hasattr(), getattr(), setattr(), delattr()



string = "123"
instance_check = isinstance(string, str)
print(instance_check)

#######  hasattr()---> used to check that particular object has particular atribute or not

##  e.g., we can chech that list object has append atribute or not...., so it help to prevent attribute error...


print()

class car:
    def __init__(self,brand):
        self.brand = brand

car_obj = car("tata")


if hasattr(car_obj,"brand"):
    print("car_obj has attribute brand")
else:
    print("not any atribute")


if hasattr(list,"append"):
    print("list has append atribute")

print()



##### getattr()---> to get data of attribute

get_attr = getattr(car_obj, "brand")
print(get_attr)       ####### output : tata


###### setattr()  ---> set value to attribute, if not create new attr of the obj

set_attr = setattr(car_obj,"model","nexon")
print(set_attr)

print(car_obj.model)  # nexon


##### delattr() ---> delete attr

del_attr= delattr(car_obj,"model")

# print(car_obj.model) # error   
```

```python
# Functional Programming
# lambda, map(), filter(), reduce() (from functools)

# eval(), exec()


#### reduce()---> take 3 args, 2 necessary, initializer is optional

from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers, 10) # Initializer of 10
print(product)
# Output: 1200 (calculated as ((((10 * 1) * 2) * 3) * 4) * 5))

print()

##### eval()---->

result = eval("5 + 3 * 2")  # Evaluates the expression "5 + 3 * 2"
print(result)  # Output: 11


######## exec()-------execute code block which is defined in string

code_block = """
a = 10
b = 20
print(a + b)
"""
exec(code_block) # Executes the statements in the string
# Output: 30
```
